##**Data Cleaning**

From Statistical desciption we see that the maximum rating is 19, which is impossible as maximum is 5. So it is understood that there is an outlier, which can be deleted.



```python
#box plot representation of outlier
playstore_data.boxplot()
```

```python
#list of data having rating greater than 5
playstore_data[playstore_data.Rating>5]
```

```python
#deleting the outlier
playstore_data.drop([10472], inplace=True)
#checking if the outlier is actually deleted
playstore_data[10470:10475]
```

From the above list of data it is understood that the outlier is deleted. And hence the row with index 10472 is deleted.

Now for data clearing, next step is to fill the null values. 

```python
# Filling the null values in Ratings column with the mean value
playstore_data['Rating'] = playstore_data['Rating'].fillna(playstore_data['Rating'].mean())
```

```python
# Filling all the null values in the playstore data
playstore_data['Type'].fillna('Free',inplace=True)
playstore_data['Content Rating'].fillna('Everyone',inplace=True)
playstore_data['Android Ver'].fillna('4.1 and up',inplace=True)
playstore_data['Current Ver'].fillna('Varies with device',inplace=True)
```

```python
# checking number of null values 
playstore_data.isnull().sum()
```

All the null values are being replaced for data analysis.

```python
#data info
playstore_data.info()
```

From the info it is clear that columns are objects which can be converted into integer or float or date-time. So we have to replace all the unwanted symbols.

```python
# Replacing symbols such as 'M', 'k', '+', ',', '$', 'Varies with device' 
playstore_data['Size'] = playstore_data['Size'].str.replace('M','e6')
playstore_data['Size'] = playstore_data['Size'].str.replace('Varies with device','0')
playstore_data['Size'] = playstore_data['Size'].str.replace('k','e3')
playstore_data['Installs'] = playstore_data['Installs'].str.replace('+','')
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',','')
playstore_data['Price'] = playstore_data['Price'].str.replace('$','')
# Checking the first 5 data
playstore_data.head()
```

```python
# Converting Size, Price and Reviews to float, and Installs to integer
playstore_data['Size'] = playstore_data['Size'].astype(float)
playstore_data['Installs'] = playstore_data['Installs'].astype(int)
playstore_data['Price'] = playstore_data['Price'].astype(float)
playstore_data['Reviews'] = playstore_data['Reviews'].astype(float)
```

```python
# checking the information in playstore dataset
playstore_data.info()
```

```python
# Check whether there are any duplicate apps in the given playstore data 
boolean = playstore_data['App'].duplicated().any()
boolean
```

```python
# List and count of number of duplicate apps
playstore_data['App'].value_counts()
```

```python
# deleting the duplicating apps
playstore_data.drop_duplicates(inplace=True)
playstore_data
```

We have deleted the row with outlier and the null values are replaced. 

Data clearing is complete!

```python
def findtop10incategory(str):
    str = str.upper()
    top10 = playstore_data[playstore_data['Category'] == str]
    top10apps = top10.sort_values(by='Installs', ascending=False).head(10)
    # Top_Apps_in_art_and_design
    plt.figure(figsize=(10,7))
    plt.title('Top 10 Installed Apps',size = 20) 
    graph = sns.barplot(x = top10apps.App, y = top10apps.Installs, palette='cubehelix')
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right')
findtop10incategory('Business')
```

```python
#word cloud imports
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

Creating word cloud with category of app

```python
#subsetting df
df_word_cloud = playstore_data[['Category','App']]
text = " ".join(word for word in df_word_cloud.App)
# Create stopword list:
stopwords = set(STOPWORDS)
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text)
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#word cloud for each category
def func_select_Category(category_name):
  df_word_cloud = playstore_data[['Category','App']]
  df_word_cloud = df_word_cloud[df_word_cloud['Category']==category_name]
  text = " ".join(word for word in df_word_cloud.App)
  # Create stopword list:
  stopwords = set(STOPWORDS)
  # Generate a word cloud image
  wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
  # Display the generated image:
  # the matplotlib way:
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```

Creating Word cloud for the category 'Communication'

```python
func_select_Category('COMMUNICATION')
```
